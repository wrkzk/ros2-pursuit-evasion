```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
df = pd.read_csv("results_2_pursuer.csv")

print(df.head())

fig, ax = plt.subplots(1, 1)
bins = np.arange(0, max(df["time_s"]), 25)
# print(bins)
ax.hist(df["time_s"], 
        bins = bins, 
        color = "lightsalmon", 
        edgecolor="white", 
        linewidth = 0.5
        )
ax.set_xlim(0, 400)

ax.set_xlabel("Time")
ax.set_ylabel("Frequency")
ax.set_title(f"Voronoi Pursuit Capture Times (2 Pursuers)")
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
plt.tight_layout()
plt.savefig("2_pursuer.png")
plt.show()
```
